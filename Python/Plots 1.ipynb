```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.arange(0,11,0.5)
y= x*x
```

```python
plt.plot(x,y)  #Functional way
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X-Y PLOT")
```

```python
#Multiple Plots
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
#Object- Oriented way
fig = plt.figure()

axis = fig.add_axes([0.2,0.5,1,1])
axis.plot(x,y)
axis.set_xlabel("X")
axis.set_ylabel("Y")
axis.set_title("MY PLOT")
```

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1,0.1,0.9,0.9])
axis2 = fig.add_axes([0.25,0.5,0.38,0.38])

axis1.plot(x,y,"r")
axis1.set_title("Bigger One")
axis1.set_xlabel("X")
axis1.set_ylabel("Y")

axis2.plot(y,x,"b")
axis2.set_title("Smaller")
axis2.set_xlabel("Y")
axis2.set_ylabel("X")
```

```python
fig,axis = plt.subplots(1,2, figsize=(8,3)) #Automatically cretes the axis array

axis[0].plot(x,y, "b", label="X vs Y")
axis[0].set_title("First")
axis[0].legend(loc=0)

axis[1].plot(y,x*x, "r", label="Y vs X Sq")
axis[1].set_title("Second")
axis[1].legend(loc=(0.2,0.8))

plt.tight_layout() #Avoids overlapping
```

```python
fig.savefig('my_plot.png', dpi=200)  #Saves plot to current directory
```
