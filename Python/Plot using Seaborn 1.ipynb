```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=True,bins=30)  #Plots distribution of the variable
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="reg") #Plot 2 variables together, kind can be a scatterplot as well
```

```python
sns.pairplot(tips, hue='sex') #Plots all varibales in the dataset
```

```python
sns.rugplot(tips['tip']) #Alternate to histogram
```

```python
sns.kdeplot(tips['tip'])
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.barplot(x='sex', y='tip', data=tips, estimator=np.std)
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='smoker',y='tip', data=tips, hue='sex') #split=True can be used to combine th ehue plots
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='smoker', dodge=True)
```

```python
sns.violinplot(x='day', y='tip', data=tips)
sns.swarmplot(x='day', y='tip', data=tips,color='black')
```

```python
#General method
sns.factorplot(x='day', y='tip', data=tips,kind='box')
```
