```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Load data
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
flights_pv=flights.pivot_table(index='month', columns='year', values='passengers')
flights_pv
```

```python
sns.heatmap(flights_pv,cmap='coolwarm', lw=1)
```

```python
sns.clustermap(flights_pv,cmap='coolwarm')
```
