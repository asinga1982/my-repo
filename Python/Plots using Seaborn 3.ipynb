```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load data
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='smoker',row='time',hue='sex', aspect=0.6, size=6)
```

```python
g=sns.PairGrid(tips) #Pairplot is a customized version of this

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#Change font by sns.set_Context('poster', font=2)

sns.set_style(style='darkgrid') # Sets background 
g = sns.FacetGrid(data=tips, row='smoker', col='time', hue='sex', size=4, aspect=1, palette='plasma')

g.map(plt.scatter, 'total_bill', 'tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,size=3,aspect=2)
```
