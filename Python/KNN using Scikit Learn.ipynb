```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Classified Data', index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(data=df, hue='TARGET CLASS')
```

```python
from sklearn.preprocessing import StandardScaler    
```

```python
scaler = StandardScaler()
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_feat = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
df_feat = pd.DataFrame(scaled_feat,columns=df.columns[:-1])
```

```python
df_feat.head()
```

```python
from sklearn.cross_validation import train_test_split
x_tr, x_ts, y_tr, y_ts = train_test_split(df_feat, df['TARGET CLASS'], test_size=0.3 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_tr,y_tr)
```

```python
pred = knn.predict(x_ts)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_ts,pred))
print(classification_report(y_ts,pred))
```

```python
error_rate =[]
for i in range(1,40):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_tr,y_tr)
        pred_i = knn.predict(x_ts)
        error_rate.append(np.mean(y_ts != pred_i))
```

```python
plt.plot(range(1,40), error_rate, ls='-',marker='o')
```

```python
#Uisng Optimal K value from elbo method
knn = KNeighborsClassifier(n_neighbors=35)
knn.fit(x_tr,y_tr)
pred = knn.predict(x_ts)
print(confusion_matrix(y_ts,pred))
print(classification_report(y_ts,pred))
```
